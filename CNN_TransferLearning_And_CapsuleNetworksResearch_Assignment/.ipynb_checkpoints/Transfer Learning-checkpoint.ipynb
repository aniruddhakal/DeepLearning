```python
import tensorflow as tf
import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV
import sklearn

import h5py


def plot_graph(history, n_epochs, save_flag=False, target_file=''):
    """
    this plot code is copied from Week 9 lecture slides of 'Deep Learning'
    """
    plt.style.use("ggplot")
#     plt.figure(dpi=150)
    plt.plot(np.arange(0, n_epochs), history.history["loss"], label="train_loss")
    plt.plot(np.arange(0, n_epochs), history.history["val_loss"], label="val_loss")
    plt.plot(np.arange(0, n_epochs), history.history["acc"], label="train_acc")
    plt.plot(np.arange(0, n_epochs), history.history["val_acc"], label="val_acc")
    plt.title("Training Loss and Accuracy")
    plt.xlabel("Epoch #")
    plt.ylabel("Loss/Accuracy")
    plt.legend()

    if not save_flag:
        plt.show()
    else:
        plt.savefig(target_file, dpi=100)


def loadDataH5(h5_file='./data1.h5'):
    with h5py.File(h5_file, 'r') as hf:
        trainX = np.array(hf.get('trainX'))
        trainY = np.array(hf.get('trainY'))
        valX = np.array(hf.get('valX'))
        valY = np.array(hf.get('valY'))
        print(trainX.shape, trainY.shape)
        print(valX.shape, valY.shape)

    return trainX, trainY, valX, valY

def load_and_preprocess():
    h5_file = './data1.h5'

    trainX, trainY, testX, testY = loadDataH5(h5_file)

    trainX = tf.keras.utils.normalize(trainX, axis=1)
    testX = tf.keras.utils.normalize(testX, axis=1)

    return trainX, trainY, testX, testY

trainX, trainY, testX, testY = load_and_preprocess()

random_seed = 171450
tf.random.set_random_seed(random_seed)
np.random.seed(random_seed)
```

```python
vgg16 = tf.keras.applications.VGG16(include_top=False, weights='imagenet', input_shape=(128, 128, 3))
vgg19 = tf.keras.applications.VGG19(include_top=False, weights='imagenet', input_shape=(128, 128, 3))
inception3 = tf.keras.applications.InceptionV3(include_top=False, weights='imagenet', input_shape=(128, 128, 3))

vgg16.summary()
```

```python
# predict weights using pre-built models
vgg16_trainX = vgg16.predict(trainX)
vgg16_testX = vgg16.predict(testX)

vgg19_trainX = vgg19.predict(trainX)
vgg19_testX = vgg19.predict(testX)

inception3_trainX = inception3.predict(trainX)
inception3_testX = inception3.predict(testX)
```

```python
# reshape
vgg16_trainX = vgg16_trainX.reshape(vgg16_trainX.shape[0], -1)
vgg16_testX = vgg16_testX.reshape(vgg16_testX.shape[0], -1)

vgg19_trainX= vgg19_trainX.reshape(vgg16_trainX.shape[0], -1)
vgg19_testX = vgg19_testX.reshape(vgg16_testX.shape[0], -1)

inception3_trainX = inception3_trainX.reshape(vgg16_trainX.shape[0], -1)
inception3_testX = inception3_testX.reshape(vgg16_testX.shape[0], -1)
```

```python
def get_random_forest_classification_accuracy(traindata, testdata, n_estimators, max_depth):
    rfc = RandomForestClassifier(n_estimators=n_estimators, random_state=random_seed, max_depth=max_depth,
                                 n_jobs=-1)
    rfc.fit(traindata, trainY)
    predictions = rfc.predict(testdata)

    accuracy = accuracy_score(testY, predictions)
    print('Random Forest Accuracy: ', accuracy_score(testY, predictions), '\n')
    # print(classification_report(testY, predictions))
    return accuracy


def get_logistic_regression_accuracy(traindata, testdata):
    rfc = sklearn.linear_model.LogisticRegression(random_state=random_seed,
                                                  solver='sag', max_iter=300)
    rfc.fit(traindata, trainY)
    predictions = rfc.predict(testdata)

    accuracy = accuracy_score(testY, predictions)
    print('Logistic Regression Accuracy: ', accuracy_score(testY, predictions), '\n')
    # print(classification_report(testY, predictions))
    return accuracy
```

```python
#max_depth=12, n_estimators=5000
# get_classification_accuracy(vgg16_trainX, vgg16_testX, 6000, 12)
get_logistic_regression_accuracy(vgg16_trainX, vgg16_testX)
```

```python
# n_estimators=[4000, 6000], depth=12
# get_classification_accuracy(vgg19_trainX, vgg19_testX, 6000, 12)
get_logistic_regression_accuracy(vgg19_trainX, vgg19_testX)
```

```python
import tensorflow as tf
from tensorflow.keras.regularizers import l2

def get_model1(pre_trained_model: tf.keras.models.Sequential, model_name, n_classes,
               INCLUDE_DROPOUT=True,
               INCLUDE_CLASSIFICATION=True):
    model = tf.keras.models.Sequential(name=model_name)
    model.add(pre_trained_model)

    model.add(tf.keras.layers.Flatten())
    model.add(tf.keras.layers.Dense(2000, activation='relu', kernel_regularizer=l2(0.05)))
    INCLUDE_DROPOUT and model.add(tf.keras.layers.Dropout(0.3))

    model.add(tf.keras.layers.Dense(550, activation='relu', kernel_regularizer=l2(0.02)))
    INCLUDE_DROPOUT and model.add(tf.keras.layers.Dropout(0.2))

    INCLUDE_CLASSIFICATION and model.add(tf.keras.layers.Dense(n_classes, activation='softmax'))

    return model
```

```python
def compile_and_fit(model, train_d1, test_d1, lr=0.01, n_epochs=5, batch_size=32):
    model.compile(loss='sparse_categorical_crossentropy',
                 optimizer=tf.keras.optimizers.SGD(lr=lr),
                 metrics=['accuracy'])
    
    history = model.fit(train_d1, trainY, epochs=n_epochs, batch_size=batch_size, validation_data=(test_d1, testY))
    
    return history
```

```python
def set_trainable(model, trainable_layers_list):
    model.trainable = True
    trainable_flag = False

    for layer in model.layers:
        if layer.name in trainable_layers_list:
#             layer.trainable = True
            trainable_flag = True

        layer.trainable = trainable_flag
        
    return model
```

```python
n_classes = 17

vgg16_model1 = get_model1(vgg16, 'vgg16+', n_classes, INCLUDE_DROPOUT=False)
vgg19_model1 = get_model1(vgg19,  'vgg19+', n_classes, INCLUDE_DROPOUT=False)
inception3_model1 = get_model1(inception3,  'Inception3+', n_classes, INCLUDE_DROPOUT=False)

print(vgg16_model1.summary())
```

```python
n_epochs = 30
learning_rate = 0.001

vgg16_model1 = get_model1(vgg16, 'vgg16+', n_classes, INCLUDE_DROPOUT=False)
history = compile_and_fit(vgg16_model1, trainX, testX, lr=learning_rate, n_epochs=n_epochs)
plot_graph(history, n_epochs)
```
